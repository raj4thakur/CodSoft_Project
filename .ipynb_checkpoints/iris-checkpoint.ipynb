```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn  as sb
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,mean_squared_error,recall_score,confusion_matrix
```

# Visualizing dataset

```python
sb.pairplot(df)
```

```python
df['species']=df['species'].map({"setosa":0,"versicolor":1,"virginica":2})
y_spec={0:"setosa",1:"versicolor",2:"virginica"}
```

```python
xtrain=df.drop(columns=["species"])
ytrain=df["species"]
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xtrain, ytrain, test_size=0.2, random_state=42,shuffle=True)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(xtrain,ytrain)
```

```python
y_predicted=model.predict(xtest)
```

```python
confusion_matrix(ytest,y_predicted)
```
